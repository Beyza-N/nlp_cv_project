```python
# news detection with logo v9 (Klavye görseli ile karşılaştırma - Türkçe açıklamalı)
import pandas as pd
from transformers import pipeline
import cv2
import numpy as np
from PIL import Image  #Pillow import

# Gerekli kütüphaneleri yükle
try:
    from PIL import Image
except ImportError:
    !pip install Pillow
try:
    import cv2
except ImportError:
    !pip install opencv-python

# --- Metin Sınıflandırma Kurulumu ---
classifier = pipeline("text-classification", model="dbmdz/bert-base-turkish-cased", return_all_scores=False)

# --- Veri Yükleme ---
csv_file_path = '/content/sample_data/haber_veri_seti/dataset.csv'
df = pd.read_csv(csv_file_path, header=None, names=['Haber Adı', 'Haber'])




def detect_logo(news_image_path, keyboard_image_path, min_match_count=10):
    """Özellik tabanlı eşleştirme (ORB) ile logo tespiti yapar."""
    try:
        # Görselleri yükle
        img1 = cv2.imread(keyboard_image_path, cv2.IMREAD_GRAYSCALE)  # Klavye (şablon)
        img2 = cv2.imread(news_image_path, cv2.IMREAD_GRAYSCALE)  # Haber

        if img1 is None or img2 is None:
            print("Hata: Görseller yüklenemedi.")
            return False

        # ORB dedektörünü ve tanımlayıcısını oluştur
        orb = cv2.ORB_create()

        # Anahtar noktaları ve tanımlayıcıları bul
        kp1, des1 = orb.detectAndCompute(img1, None)  # Klavye (şablon)
        kp2, des2 = orb.detectAndCompute(img2, None)  # Haber

        if des1 is None or des2 is None:
           print("Hata: Yeterli anahtar nokta bulunamadı.")
           return False
        # BFMatcher (Brute-Force Matcher) oluştur
        bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

        # Eşleşmeleri bul
        matches = bf.match(des1, des2)

        # Eşleşmeleri mesafeye göre sırala (en iyi eşleşmeler önce)
        matches = sorted(matches, key=lambda x: x.distance)

        # Yeterli sayıda iyi eşleşme var mı?
        if len(matches) > min_match_count:
            print(f"Yeterli eşleşme bulundu: {len(matches)}/{min_match_count}")
            return True  # Logo bulundu
        else:
            print(f"Yeterli eşleşme bulunamadı: {len(matches)}/{min_match_count}")
            return False  # Logo bulunamadı


    except Exception as e:
        print(f"Logo tespiti sırasında hata: {e}")
        return False


def classify_and_label(haber_baslik, haber_metni, sirket, keywords, news_image_path, keyboard_image_path):
    """Metin ve logo tespiti ile haberleri sınıflandırır."""
    combined_text = f"{haber_baslik} {haber_metni}".upper()

    # Metin Tabanlı Kontrol
    if sirket.upper() in combined_text and any(keyword.upper() in combined_text for keyword in keywords):
        try:
            result = classifier(haber_metni[:512])
            label = result[0]['label']
            if label == "LABEL_1":
                return "ilgili"
        except Exception as e:
            print(f"Model hatası: {e}")
            return "ilgisiz"

    # Logo Tabanlı Kontrol
    if detect_logo(news_image_path, keyboard_image_path, threshold=0.7):
        return "ilgili"

    return "ilgisiz"



# --- Ana Yürütme ---
# --- Ana Yürütme ---
sirket_adi = input("Lütfen şirket adını girin: ").strip()
anahtar_kelime_input = input("Lütfen anahtar kelimeleri virgülle ayırarak girin: ")
anahtar_kelimeler = [kelime.strip() for kelime in anahtar_kelime_input.split(',')]

# Varsayılan haber görseli yolu (örnek - ilk görselin yolu)
# Bunu, veri setinizdeki GERÇEK bir görselin yolu ile değiştirin!
default_news_image_path = "/content/sample_data/haber_veri_seti/logos/images.jpeg"

news_image_path = input(f"Lütfen haber görselinin dosya yolunu girin (varsayılan: {default_news_image_path}): ") or default_news_image_path
keyboard_image_path = input("Lütfen klavyedeki görselin dosya yolunu girin: ")

# İlk haber için örnek kullanım (isterseniz döngüde kullanabilirsiniz)
first_row = df.iloc[0]
haber_baslik = str(first_row['Haber Adı'])
haber_metni = str(first_row['Haber'])

durum = classify_and_label(haber_baslik, haber_metni, sirket_adi, anahtar_kelimeler, news_image_path, keyboard_image_path)
print(f"Haber durumu: {durum}")

# İsteğe bağlı: Diğer haberler için döngü
# for index, row in df.iterrows():
#     haber_baslik = str(row['Haber Adı'])
#     haber_metni = str(row['Haber'])
#     durum = classify_and_label(haber_baslik, haber_metni, sirket_adi, anahtar_kelimeler, news_image_path, keyboard_image_path)
#     print(f"Haber {index+1} durumu: {durum}")
```

```python
# news detection with logo v14 (Tek referans logo, KLAVYE yüklemesi, SIFT, Lowe's - Türkçe açıklamalı)
import pandas as pd
from transformers import pipeline
import cv2
import numpy as np
import os
from PIL import Image

# Gerekli kütüphaneleri yükle
try:
    from PIL import Image
except ImportError:
    !pip install Pillow
try:
    import cv2
except ImportError:
    !pip install opencv-python
try:
    import cv2.xfeatures2d  # SIFT için
except ImportError:
    !pip install opencv-contrib-python --user

# --- Metin Sınıflandırma Kurulumu ---
classifier = pipeline("text-classification", model="dbmdz/bert-base-turkish-cased", return_all_scores=False)

# --- Veri Yükleme ---
csv_file_path = '/content/sample_data/haber_veri_seti/dataset.csv'
df = pd.read_csv(csv_file_path, header=None, names=['Haber Adı', 'Haber'])

# --- Referans Logo Yolu ---
REFERENCE_LOGO_PATH = '/content/sample_data/haber_veri_seti/logos/images1.jpeg'  # Referans logonuzun yolu


def detect_logo(news_image_path, keyboard_image_path):

    try:
        # Haber görselini yükle
        news_img = cv2.imread(news_image_path, cv2.IMREAD_GRAYSCALE)
        if news_img is None:
            print(f"Hata: Haber görseli {news_image_path} yüklenemedi.")
            return False

        # Referans logoyu yükle
        reference_logo_img = cv2.imread(REFERENCE_LOGO_PATH, cv2.IMREAD_GRAYSCALE)
        if reference_logo_img is None:
            print(f"Hata: Referans logo {REFERENCE_LOGO_PATH} yüklenemedi.")
            return False

        # Klavye logosunu (yüklenen logoyu) yükle
        keyboard_img = cv2.imread(keyboard_image_path, cv2.IMREAD_GRAYSCALE)
        if keyboard_img is None:
            print("Hata: Klavye (yüklenen) logo görseli yüklenemedi.")
            return False

        # SIFT dedektörünü oluştur
        sift = cv2.SIFT_create()

        # --- Referans Logo <-> Haber Görseli Eşleşmesi ---
        kp_ref, des_ref = sift.detectAndCompute(reference_logo_img, None)  # Referans logo
        kp_news, des_news = sift.detectAndCompute(news_img, None)  # Haber görseli
        if des_ref is None or des_news is None:
            print("Hata: Referans logo veya haber görselinde yeterli özellik bulunamadı.")
            return False

        bf_ref_news = cv2.BFMatcher()
        matches_ref_news = bf_ref_news.knnMatch(des_ref, des_news, k=2)

        good_matches_ref_news = []
        for m, n in matches_ref_news:
            if m.distance < 0.75 * n.distance:
                good_matches_ref_news.append(m)

        min_match_count = 10
        news_logo_match = len(good_matches_ref_news) > min_match_count  # Haber görselinde logo bulundu mu?

        if not news_logo_match:
            print(f"Haber görselinde referans logo bulunamadı: {len(good_matches_ref_news)}/{min_match_count}")
            return False  # Haber görselinde referans logo yoksa, gerisine bakmaya gerek yok
        else:
            print(f"Haber görselinde referans logo bulundu: {len(good_matches_ref_news)}/{min_match_count}")

        # --- Referans Logo <-> Klavye Logosu Eşleşmesi ---
        kp_ref, des_ref = sift.detectAndCompute(reference_logo_img, None) #Referans
        kp_key, des_key = sift.detectAndCompute(keyboard_img, None)  # Klavye
        if des_ref is None or des_key is None:
            print("Hata: Referans logo veya klavye logosunda yeterli özellik bulunamadı.")
            return False

        bf_ref_key = cv2.BFMatcher()
        matches_ref_key = bf_ref_key.knnMatch(des_ref, des_key, k=2)

        good_matches_ref_key = []
        for m, n in matches_ref_key:
            if m.distance < 0.75 * n.distance:
                good_matches_ref_key.append(m)


        if len(good_matches_ref_key) > min_match_count:
            print(f"Referans ve klavyeden yüklenen logo eşleşti: {len(good_matches_ref_key)}/{min_match_count}")
            return True  # Referans ve klavye logoları eşleşti
        else:
            print(f"Referans ve klavyeden yüklenen logo eşleşmedi: {len(good_matches_ref_key)}/{min_match_count}")
            return False  # Referans ve klavye logoları eşleşmedi


    except Exception as e:
        print(f"Logo tespiti sırasında hata: {e}")
        return False



def classify_and_label(haber_baslik, haber_metni, target_company, keywords, news_image_path, keyboard_image_path):
    """Metin ve logo tespiti ile haberleri sınıflandırır."""
    combined_text = f"{haber_baslik} {haber_metni}".upper()

    # --- Metin Tabanlı Kontrol ---
    text_match = False
    text_result_message = "Metinde eşleşme bulunamadı."
    if target_company.upper() in combined_text and any(keyword.upper() in combined_text for keyword in keywords):
        try:
            result = classifier(haber_metni[:512])
            label = result[0]['label']
            if label == "LABEL_1":
                text_match = True
                text_result_message = f"Metinde eşleşme bulundu ({target_company} ve anahtar kelimeler)."
        except Exception as e:
            print(f"Metin sınıflandırma hatası: {e}")
            text_result_message = "Metin sınıflandırması sırasında hata oluştu."

    # --- Logo Tabanlı Kontrol ---
    logo_match = detect_logo(news_image_path, keyboard_image_path)

    # --- Sonuçları Göster ---
    print(text_result_message)

    if text_match or logo_match:
        return "ilgili"
    else:
        return "ilgisiz"



# --- Ana Yürütme ---
sirket_adi = input("Lütfen şirket adını girin: ").strip()
anahtar_kelime_input = input("Lütfen anahtar kelimeleri virgülle ayırarak girin: ")
anahtar_kelimeler = [kelime.strip() for kelime in anahtar_kelime_input.split(',')]

# Varsayılan haber görseli yolu
default_news_image_path = "/content/sample_data/haber_veri_seti/logos/images1.jpeg"  # Kendi yolunuzla DEĞİŞTİRİN!

news_image_path = input(f"Lütfen haber görselinin dosya yolunu girin (varsayılan: {default_news_image_path}): ") or default_news_image_path
keyboard_image_path = input("Lütfen klavyedeki (yüklenecek) logonun dosya yolunu girin: ")


# İlk haber için örnek kullanım
first_row = df.iloc[0]
haber_baslik = str(first_row['Haber Adı'])
haber_metni = str(first_row['Haber'])

durum = classify_and_label(haber_baslik, haber_metni, sirket_adi, anahtar_kelimeler, news_image_path, keyboard_image_path)
print(f"Haber durumu: {durum}")

# İsteğe bağlı: Diğer haberler için döngü
# for index, row in df.iterrows():
#     haber_baslik = str(row['Haber Adı'])
#     haber_metni = str(row['Haber'])
#     durum = classify_and_label(haber_baslik, haber_metni, sirket_adi, anahtar_kelimeler, news_image_path, keyboard_image_path)
#     print(f"Haber {index + 1} durumu: {durum}")
```

```python
# news detection with logo v24 (Özet sonuçlar, metin/logo ayrı, VEYA - Türkçe, KESİN)
import pandas as pd
from transformers import pipeline
import cv2
import numpy as np
import os
from PIL import Image

# --- Kütüphaneler ---
try:
    from PIL import Image
except ImportError:
    !pip install Pillow
try:
    import cv2
except ImportError:
    !pip install opencv-python
try:
    import cv2.xfeatures2d
except ImportError:
    !pip install opencv-contrib-python --user

# --- Metin Sınıflandırma ---
classifier = pipeline("text-classification", model="dbmdz/bert-base-turkish-cased", return_all_scores=False)

# --- Veri Yükleme ---
csv_file_path = '/content/sample_data/haber_veri_seti/dataset.csv'
df = pd.read_csv(csv_file_path, names=['Haber Adı', 'Haber'])  # Sadece 2 sütun

# --- Logo Klasörü (REFERANS logolar burada) ---
LOGO_DIR = '/content/sample_data/haber_veri_seti/logos/'

# --- Haber Veri Seti Ana Dizini ---
DATASET_BASE_DIR = '/content/sample_data/haber_veri_seti/'


def detect_logo(news_image_path, keyboard_image_path, logo_dir):
    """Özellik tabanlı eşleştirme (SIFT) ile logo tespiti."""
    try:
        news_img = cv2.imread(news_image_path, cv2.IMREAD_GRAYSCALE)
        keyboard_img = cv2.imread(keyboard_image_path, cv2.IMREAD_GRAYSCALE)
        if news_img is None or keyboard_img is None:
            print("Hata: Haber veya klavye görseli yüklenemedi.")
            return False

        sift = cv2.SIFT_create()
        kp_news, des_news = sift.detectAndCompute(news_img, None)
        if des_news is None:
            print("Hata: Haber görselinde yeterli özellik bulunamadı.")
            return False

        best_logo_path = None
        best_match_count = -1

        for logo_filename in os.listdir(logo_dir):
            if logo_filename.endswith(('.png', '.jpg', '.jpeg')):
                logo_path = os.path.join(logo_dir, logo_filename)
                try:
                    logo_img = cv2.imread(logo_path, cv2.IMREAD_GRAYSCALE)
                    if logo_img is None:
                        print(f"Uyarı: {logo_path} yüklenemedi.")
                        continue

                    kp_logo, des_logo = sift.detectAndCompute(logo_img, None)
                    if des_logo is None:
                        print(f"Uyarı: {logo_path} logosunda yeterli özellik yok.")
                        continue

                    bf = cv2.BFMatcher()
                    matches = bf.knnMatch(des_logo, des_news, k=2)
                    good_matches = []
                    for m, n in matches:
                        if m.distance < 0.75 * n.distance:
                            good_matches.append(m)

                    if len(good_matches) > best_match_count:
                        best_match_count = len(good_matches)
                        best_logo_path = logo_path

                except Exception as e:
                    print(f"{logo_path} işlenirken hata: {e}")
                    continue

        min_match_count = 10
        if best_match_count < min_match_count:
            # print(f"Haber görselinde logo klasöründeki hiçbir logoyla eşleşme yok: {best_match_count}/{min_match_count}")
            return False  # Haber görseli eşleşmedi

        # print(f"Haber görseli eşleşti: {best_logo_path} ({best_match_count}/{min_match_count})")
        best_logo_img = cv2.imread(best_logo_path, cv2.IMREAD_GRAYSCALE)
        kp_best, des_best = sift.detectAndCompute(best_logo_img, None)
        kp_key, des_key = sift.detectAndCompute(keyboard_img, None)
        if des_best is None or des_key is None:
            print("Hata: Referans veya klavye logosunda yeterli özellik yok.")
            return False

        bf = cv2.BFMatcher()
        matches = bf.knnMatch(des_best, des_key, k=2)
        good_matches = []
        for m, n in matches:
            if m.distance < 0.75 * n.distance:
                good_matches.append(m)

        if len(good_matches) > min_match_count:
            # print(f"Referans ve klavye logoları eşleşti: {len(good_matches)}/{min_match_count}")
            return True
        else:
            # print(f"Referans ve klavye logoları EŞLEŞMEDİ: {len(good_matches)}/{min_match_count}")
            return False

    except Exception as e:
        print(f"Logo tespiti sırasında hata: {e}")
        return False



def analyze_news(haber_baslik, haber_metni, target_company, keywords, news_image_dir, keyboard_image_path):
    """Tek bir haber için metin ve logo analizini yapar ve sonuçları (True/False) döndürür."""

    # --- Metin Analizi ---
    combined_text = f"{haber_baslik} {haber_metni}".upper()
    text_match = False
    if target_company.upper() in combined_text and any(keyword.upper() in combined_text for keyword in keywords):
        try:
            result = classifier(haber_metni[:512])
            if result[0]['label'] == "LABEL_1":
                text_match = True
        except Exception as e:
            print(f"Metin sınıflandırma hatası: {e}")

    # --- Haber Görselinin Tam Yolunu Oluştur ---
    news_image_path = os.path.join(news_image_dir, "gorsel_1.jpg")  # Varsayılan ilk görsel.  DİKKAT!
    # news_image_path = os.path.join(DATASET_BASE_DIR, news_image_dir) #önceden böyleydi
    try:
        if not os.path.exists(news_image_path):
            # print(f"Hata: {news_image_path} dosyası bulunamadı.") #Artık yazdırmıyoruz
            logo_match = False
        else:
            # print(f"Haber görseli: {news_image_path}") #Artık yazdırmıyoruz.
            # --- Logo Analizi ---
            logo_match = detect_logo(news_image_path, keyboard_image_path, LOGO_DIR)
    except FileNotFoundError:
        print(f"Hata: {news_image_path} dizini bulunamadı.")
        logo_match = False

    return text_match, logo_match  # SONUÇLARI DÖNDÜR


# --- Ana Yürütme ---
sirket_adi = input("Lütfen şirket adını girin: ").strip()
anahtar_kelime_input = input("Lütfen anahtar kelimeleri virgülle ayırarak girin: ")
anahtar_kelimeler = [kelime.strip() for kelime in anahtar_kelime_input.split(',')]
keyboard_image_path = input("Lütfen klavyedeki (yüklenecek) logonun dosya yolunu girin: ")

# Tüm haberler için sonuçları saklayacak listeler
all_results = []

# Tüm haberleri analiz et
for index, row in df.iterrows():
    haber_baslik = str(row['Haber Adı'])
    haber_metni = str(row['Haber'])
    haber_gorsel_dir = os.path.join(DATASET_BASE_DIR, f"haber_{index + 1}")  # Dizin adını oluştur

    text_match, logo_match = analyze_news(haber_baslik, haber_metni, sirket_adi, anahtar_kelimeler, haber_gorsel_dir, keyboard_image_path)

    all_results.append({
        'haber_index': index + 1,
        'haber_baslik': haber_baslik,
        'text_match': text_match,
        'logo_match': logo_match,
        'overall_match': text_match or logo_match  # Herhangi biri True ise True
    })

# --- Sonuçları Özet Olarak Yazdır ---
print("\n--- Sonuçlar ---")
print(f"Toplam haber sayısı: {len(all_results)}")

num_overall_positive = 0  # Toplam "ilgili" haber sayısı

for result in all_results:
    if result['overall_match']:
        print(f"Haber {result['haber_index']} ({result['haber_baslik'][:30]}...): İLGİLİ")  # İlk 30 karakter
        num_overall_positive += 1


print(f"\nToplam İLGİLİ haber sayısı: {num_overall_positive}/{len(all_results)}")
```

```python
# news detection with logo v28 (ÇOKLU referans logo, klavye, SIFT, Lowe's, hedef şirket - Türkçe, KESİN ÇÖZÜM)  logoları görmüyor hala, ama metin iyi sonuç ve çıktı yapısı istediğim gibi
import pandas as pd
from transformers import pipeline
import cv2
import numpy as np
import os
from PIL import Image

# --- Kütüphaneler ---
try:
    from PIL import Image
except ImportError:
    !pip install Pillow
try:
    import cv2
except ImportError:
    !pip install opencv-python
try:
    import cv2.xfeatures2d
except ImportError:
    !pip install opencv-contrib-python --user

# --- Metin Sınıflandırma ---
classifier = pipeline("text-classification", model="dbmdz/bert-base-turkish-cased", return_all_scores=False)

# --- Veri Yükleme ---
csv_file_path = '/content/sample_data/haber_veri_seti/dataset.csv'
df = pd.read_csv(csv_file_path, names=['Haber Adı', 'Haber'])  # Sadece 2 sütun

# --- Logo Klasörü (REFERANS logolar burada) ---
LOGO_DIR = '/content/sample_data/haber_veri_seti/logos/'

# --- Haber Veri Seti Ana Dizini ---
DATASET_BASE_DIR = '/content/sample_data/haber_veri_seti/'


def detect_logo(news_image_path, keyboard_image_path, logo_dir):
    """Özellik tabanlı eşleştirme (SIFT) ile logo tespiti."""
    try:
        # Haber ve klavye görsellerini yükle
        news_img = cv2.imread(news_image_path, cv2.IMREAD_GRAYSCALE)
        keyboard_img = cv2.imread(keyboard_image_path, cv2.IMREAD_GRAYSCALE)
        if news_img is None or keyboard_img is None:
            print("Hata: Haber veya klavye görseli yüklenemedi.")
            return False

        # SIFT dedektörünü oluştur
        sift = cv2.SIFT_create()

        # --- 1. Aşama: Haber Görseli <-> Logo Klasöründeki TÜM Logolar ---

        # Haber görseli için keypoint ve descriptor'ları hesapla
        kp_news, des_news = sift.detectAndCompute(news_img, None)
        if des_news is None:
            print("Hata: Haber görselinde yeterli özellik bulunamadı.")
            return False

        best_logo_path = None  # En iyi eşleşen logonun YOLU
        best_match_count = -1

        for logo_filename in os.listdir(logo_dir):
            if logo_filename.endswith(('.png', '.jpg', '.jpeg')):
                logo_path = os.path.join(logo_dir, logo_filename)
                try:
                    logo_img = cv2.imread(logo_path, cv2.IMREAD_GRAYSCALE)
                    if logo_img is None:
                        print(f"Uyarı: {logo_path} yüklenemedi.")
                        continue

                    # Logo klasöründeki HER BİR logo için keypoint ve descriptor'ları hesapla
                    kp_logo, des_logo = sift.detectAndCompute(logo_img, None)
                    if des_logo is None:
                        print(f"Uyarı: {logo_path} logosunda yeterli özellik yok.")
                        continue

                    # Haber görseli ile logo klasöründeki logoyu karşılaştır
                    bf = cv2.BFMatcher()
                    matches = bf.knnMatch(des_logo, des_news, k=2)  # DOĞRU SIRA: des_logo (sorgu), des_news (eğitim)

                    good_matches = []
                    for m, n in matches:
                        if m.distance < 0.75 * n.distance:
                            good_matches.append(m)

                    # En iyi eşleşmeyi güncelle
                    if len(good_matches) > best_match_count:
                        best_match_count = len(good_matches)
                        best_logo_path = logo_path  # En iyi eşleşen logonun YOLU

                except Exception as e:
                    print(f"{logo_path} işlenirken hata: {e}")
                    continue

        # Haber görselinde, logo klasöründeki herhangi bir logoyla eşleşme var mı?
        min_match_count = 10
        if best_match_count < min_match_count:
            print(f"Haber görselinde, logo klasöründeki ({logo_dir}) hiçbir logoyla yeterli eşleşme bulunamadı. En iyi: {best_match_count}/{min_match_count}")
            return False  # Eşleşme yok

        print(f"Haber görseli ile logo klasöründeki en iyi eşleşen logo: {best_logo_path} ({best_match_count}/{min_match_count})")

        # --- 2. Aşama: Klavye Logosu <-> Haber Görselinde BULUNAN Logo ---

        # Şimdi, haber görselinde BULDUĞUMUZ logo (best_logo_path) ile
        # klavyeden yüklenen logoyu (keyboard_img) karşılaştıracağız.

        # En iyi eşleşen logoyu (haber görselinde bulduğumuz) tekrar yükle
        best_logo_img = cv2.imread(best_logo_path, cv2.IMREAD_GRAYSCALE)
        if best_logo_img is None:
            print("Hata: En iyi eşleşen logo dosyası tekrar yüklenemedi.")
            return False

        # En iyi eşleşen logo (best_logo_img) ve klavye logosu (keyboard_img) için
        # keypoint ve descriptor'ları hesapla
        kp_best, des_best = sift.detectAndCompute(best_logo_img, None)  # En iyi eşleşen (haberden)
        kp_key, des_key = sift.detectAndCompute(keyboard_img, None)  # Klavye
        if des_best is None or des_key is None:
            print("Hata: En iyi eşleşen logo veya klavye logosunda yeterli özellik yok.")
            return False

        # Eşleştirme
        bf_key = cv2.BFMatcher()
        matches_key = bf_key.knnMatch(des_best, des_key, k=2)  # En iyi eşleşen <-> Klavye (DOĞRU SIRA)

        good_matches_key = []
        for m, n in matches_key:
            if m.distance < 0.75 * n.distance:
                good_matches_key.append(m)

        # Yeterli eşleşme var mı?
        if len(good_matches_key) > min_match_count:
            print(f"Logo klasöründeki logo ile klavyeden yüklenen logo eşleşti: {len(good_matches_key)}/{min_match_count}")
            return True  # Eşleşme VAR
        else:
            print(f"Logo klasöründeki logo ile klavyeden yüklenen logo eşleşmedi: {len(good_matches_key)}/{min_match_count}")
            return False  # Eşleşme YOK

    except Exception as e:
        print(f"Logo tespiti sırasında hata: {e}")
        return False


def analyze_news(haber_baslik, haber_metni, target_company, keywords, news_image_dir, keyboard_image_path):
    """Tek bir haber için metin ve logo analizini yapar ve sonuçları (True/False) döndürür."""

    # --- Metin Analizi ---
    combined_text = f"{haber_baslik} {haber_metni}".upper()
    text_match = False
    if target_company.upper() in combined_text and any(keyword.upper() in combined_text for keyword in keywords):
        try:
            result = classifier(haber_metni[:512])
            if result[0]['label'] == "LABEL_1":
                text_match = True
        except Exception as e:
            print(f"Metin sınıflandırma hatası: {e}")

    # --- Haber Görselinin Tam Yolunu Oluştur ---
    news_image_path = os.path.join(news_image_dir, "gorsel_1.jpg")  # Varsayılan ilk görsel
    try:
        if not os.path.exists(news_image_path):
            logo_match = False
        else:
            # --- Logo Analizi ---
            logo_match = detect_logo(news_image_path, keyboard_image_path, LOGO_DIR)
    except FileNotFoundError:
        print(f"Hata: {news_image_path} dizini bulunamadı.")
        logo_match = False

    return text_match, logo_match


# --- Ana Yürütme ---
sirket_adi = input("Lütfen şirket adını girin: ").strip()
anahtar_kelime_input = input("Lütfen anahtar kelimeleri virgülle ayırarak girin: ")
anahtar_kelimeler = [kelime.strip() for kelime in anahtar_kelime_input.split(',')]
keyboard_image_path = input("Lütfen klavyedeki (yüklenecek) logonun dosya yolunu girin: ")

# Tüm haberler için sonuçları sakla
num_text_positive = 0
num_logo_positive = 0
num_overall_positive = 0

# Tüm haberleri analiz et
for index, row in df.iterrows():
    haber_baslik = str(row['Haber Adı'])
    haber_metni = str(row['Haber'])
    haber_gorsel_dir = os.path.join(DATASET_BASE_DIR, f"haber_{index + 1}")

    text_match, logo_match = analyze_news(haber_baslik, haber_metni, sirket_adi, anahtar_kelimeler, haber_gorsel_dir, keyboard_image_path)

    # Sonuçları güncelle
    if text_match:
        num_text_positive += 1
    if logo_match:
        num_logo_positive += 1
    if text_match or logo_match:
        num_overall_positive += 1

# --- Sonuçları Özet Olarak Yazdır ---
print("\n--- Sonuçlar ---")
print(f"Toplam Haber Sayısı: {len(df)}")
print(f"Metin İçin İlgili Haber Sayısı: {num_text_positive}")
print(f"Logo İçin İlgili Haber Sayısı: {num_logo_positive}")
print(f"Toplam İlgili Haber Sayısı (Metin VEYA Logo): {num_overall_positive}")
```

```python
import cv2
import numpy as np
import os  # os modülünü ekle
from PIL import Image

def detect_logo(news_image_path, keyboard_image_path, reference_logo_path):
    """SIFT ile logo tespiti (klavye ve referans logo)."""
    try:
        # Haber, klavye ve referans logoyu yükle
        news_img = cv2.imread(news_image_path, cv2.IMREAD_GRAYSCALE)
        keyboard_img = cv2.imread(keyboard_image_path, cv2.IMREAD_GRAYSCALE)
        ref_logo_img = cv2.imread(reference_logo_path, cv2.IMREAD_GRAYSCALE)

        if news_img is None or keyboard_img is None or ref_logo_img is None:
            print("Hata: Görseller yüklenemedi.")
            return False

        # SIFT dedektörünü oluştur
        sift = cv2.SIFT_create()

        # Keypoint ve descriptor'ları bul (Referans <-> Haber)
        kp_ref, des_ref = sift.detectAndCompute(ref_logo_img, None)  # Referans
        kp_news, des_news = sift.detectAndCompute(news_img, None)  # Haber
        if des_ref is None or des_news is None:
            print("Hata: Referans veya haber görselinde yeterli özellik yok.")
            return False

        bf = cv2.BFMatcher()
        matches = bf.knnMatch(des_ref, des_news, k=2)  # Referans <-> Haber

        good_matches = []
        for m, n in matches:
            if m.distance < 0.75 * n.distance:
                good_matches.append(m)

        min_match_count = 10
        if len(good_matches) < min_match_count:
            print(f"Haber görseli ve referans logo eşleşmedi: {len(good_matches)}/{min_match_count}")
            return False
        print(f"Haber görseli ve referans logo eşleşti: {len(good_matches)}/{min_match_count}")


        # Keypoint ve descriptor'ları bul (Klavye <-> Referans)
        kp_ref, des_ref = sift.detectAndCompute(ref_logo_img, None) #Referans
        kp_key, des_key = sift.detectAndCompute(keyboard_img, None)  # Klavye
        if des_ref is None or des_key is None:
          print("Referans veya klavye logosunda yeterli özellik bulunamadı")
          return False

        bf_key = cv2.BFMatcher()
        matches_key = bf_key.knnMatch(des_ref, des_key, k=2)  # Referans <-> Klavye

        good_matches_key = []
        for m, n in matches_key:
            if m.distance < 0.75 * n.distance:
                good_matches_key.append(m)

        if len(good_matches_key) > min_match_count:
            print(f"Klavye ve referans logoları eşleşti: {len(good_matches_key)}/{min_match_count}")
            return True
        else:
            print(f"Klavye ve referans logoları eşleşmedi: {len(good_matches_key)}/{min_match_count}")
            return False

    except Exception as e:
        print(f"Hata: {e}")
        return False

# --- TEST ---
# Kendi dosya yollarınızla GÜNCELLEYİN:
news_image_path = '/content/pagani.jpg'  # Örnek HABER görseli
reference_logo_path = '/content/sample_data/haber_veri_seti/logos/'  # Örnek REFERANS logo

keyboard_image_path = input("Lütfen klavyedeki logonun yolunu girin: ")

detect_logo(news_image_path, keyboard_image_path, reference_logo_path)
```

```python
# news_detection_v36_final (ÇOKLU referans logo, klavye VS LOGO FOLDER, BASİT, KESİN ÇÖZÜM - FIXED LOOP)   LOGO İÇİN ÇALIŞIYOR
import pandas as pd
from transformers import pipeline
import cv2
import numpy as np
import os
from PIL import Image

# --- Kütüphaneler ---
try:
    from PIL import Image
except ImportError:
    !pip install Pillow
try:
    import cv2
except ImportError:
    !pip install opencv-python
try:
    import cv2.xfeatures2d
except ImportError:
    !pip install opencv-contrib-python --user

# --- Metin Sınıflandırma ---
classifier = pipeline("text-classification", model="dbmdz/bert-base-turkish-cased", return_all_scores=False)

# --- Veri Yükleme ---
csv_file_path = '/content/sample_data/haber_veri_seti/dataset.csv'
df = pd.read_csv(csv_file_path, names=['Haber Adı', 'Haber'])  # Sadece 2 sütun

# --- Logo Klasörü (REFERANS logolar burada) ---
LOGO_DIR = '/content/sample_data/haber_veri_seti/logos/'

def detect_logo(keyboard_image_path, logo_dir):
    """Özellik tabanlı eşleştirme (SIFT) ile logo tespiti.

    Klavyeden yüklenen logoyu, logo_dir'deki TÜM logolarla karşılaştırır.
    Returns the count of matched logos from logo_dir.
    """
    try:
        # Klavye görselini yükle (GRAYSCALE)
        keyboard_img = cv2.imread(keyboard_image_path, cv2.IMREAD_GRAYSCALE)
        if keyboard_img is None:
            print("Hata: Klavye görseli yüklenemedi.")
            return 0  # Return 0 instead of False

        # SIFT dedektörünü oluştur
        sift = cv2.SIFT_create()
        kp_key, des_key = sift.detectAndCompute(keyboard_img, None) #Klavyeden yüklenen logo özellikleri
        if des_key is None:
            print("Hata: Klavye görselinde yeterli özellik bulunamadı.")
            return 0 # Return 0 instead of False

        min_match_count = 10
        matched_logo_count = 0 # Sayacı başlat

        # --- Klavye Logosu <-> Logo Klasöründeki TÜM Logolar ---
        for logo_filename in os.listdir(logo_dir):
            if logo_filename.endswith(('.png', '.jpg', '.jpeg')):
                logo_path = os.path.join(logo_dir, logo_filename)
                try:
                    logo_img = cv2.imread(logo_path, cv2.IMREAD_GRAYSCALE)
                    if logo_img is None:
                        print(f"Uyarı: {logo_path} yüklenemedi.")
                        continue

                    # Logo klasöründeki HER BİR logo için keypoint/descriptor
                    kp_logo, des_logo = sift.detectAndCompute(logo_img, None)
                    if des_logo is None:
                        print(f"Uyarı: {logo_path} logosunda yeterli özellik yok.")
                        continue

                    # Klavye logosu ile logo klasöründeki logoyu karşılaştır
                    bf = cv2.BFMatcher()
                    matches = bf.knnMatch(des_key, des_logo, k=2)  # Keyboard -> Logo klasörü (KARŞILAŞTIRMA)

                    good_matches = []
                    for m, n in matches:
                        if m.distance < 0.75 * n.distance:
                            good_matches.append(m)

                    if len(good_matches) > min_match_count:
                        print(f"Klavyeden yüklenen logo, logo klasöründeki {logo_filename} ile eşleşti: {len(good_matches)}/{min_match_count}")
                        matched_logo_count += 1 # Sayacı artır
                        # break  # Artık döngüden çıkmaya gerek yok, tüm logoları saymak istiyoruz

                except Exception as e:
                    print(f"Hata: {logo_path} işlenirken hata: {e}")
                    continue


        if matched_logo_count > 0:
            print(f"Klavyeden yüklenen logo, logo klasöründeki {matched_logo_count} logo ile eşleşti.")
            return matched_logo_count # Eşleşen logo sayısını döndür
        else:
            print(f"Klavyeden yüklenen logo, logo klasöründeki hiçbir logoyla yeterli eşleşme bulunamadı.")
            return 0 # Eşleşme YOK, 0 döndür

    except Exception as e:
        print(f"Logo tespiti sırasında hata: {e}")
        return 0 # Hata durumunda 0 döndür


def classify_and_label(haber_baslik, haber_metni, target_company, keywords): #news_image_path, keyboard_image_path KALDIRILDI
    """Metin ve logo tespiti ile haberleri sınıflandırır."""

    # --- Metin Analizi ---
    combined_text = f"{haber_baslik} {haber_metni}".upper()
    text_match = False
    if target_company.upper() in combined_text and any(keyword.upper() in combined_text for keyword in keywords):
        try:
            result = classifier(haber_metni[:512])
            if result[0]['label'] == "LABEL_1":
                text_match = True
        except Exception as e:
            print(f"Metin sınıflandırma hatası: {e}")

    return text_match # Sadece metin sonucu döndür


# --- Ana Yürütme ---
sirket_adi = input("Lütfen şirket adını girin: ").strip().upper()
anahtar_kelime_input = input("Lütfen anahtar kelimeleri virgülle ayırarak girin: ")
anahtar_kelimeler = [kelime.strip().upper() for kelime in anahtar_kelime_input.split(',')]
keyboard_image_path = input("Lütfen klavyedeki (yüklenecek) logonun dosya yolunu girin: ")

# Logo analizi (dataset döngüsünden ÖNCE, SADECE BİR KEZ ÇAĞRILIR)
num_logo_positive = detect_logo(keyboard_image_path, LOGO_DIR) # Doğru logo sayısını al

# Sonuçları sakla
num_text_positive = 0
num_overall_positive = 0

# Haberleri analiz et (dataset döngüsü, SADECE METİN ANALİZİ İÇİN)
for index, row in df.iterrows():
    haber_baslik = str(row['Haber Adı'])
    haber_metni = str(row['Haber'])

    text_match = classify_and_label(haber_baslik, haber_metni, sirket_adi, anahtar_kelimeler) # Sadece metin analizi

    # Metin sonuçlarını güncelle
    if text_match:
        num_text_positive += 1

# Genel sonuçları hesapla (METİN veya LOGO)
num_overall_positive = num_text_positive + num_logo_positive # Metin ve logo sayılarını TOPLA (VEYA)

# --- Sonuçları Yazdır ---
print("\n--- Sonuçlar ---")
print(f"Toplam Haber Sayısı: {len(df)}")
print(f"Metin İçin İlgili Haber Sayısı ({sirket_adi} ve Anahtar Kelimeler): {num_text_positive}")
print(f"Logo İçin İlgili Haber Sayısı: {num_logo_positive}") # Doğru logo sayısı
print(f"Toplam İlgili Haber Sayısı (Metin VEYA Logo): {num_overall_positive}") # Genel toplam
```
